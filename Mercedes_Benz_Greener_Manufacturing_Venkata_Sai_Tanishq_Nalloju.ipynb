###Dropping Zero variance columns

```python
'''
Dropping Zero variance columns
'''
from sklearn.feature_selection import VarianceThreshold
constant_filter = VarianceThreshold(threshold=0)
constant_filter.fit(train_features.iloc[:, 8:])
len(train_features.iloc[:, 8:].columns[constant_filter.get_support()])

constant_columns = [column for column in train_features.iloc[:, 8:].columns
                    if column not in train_features.iloc[:, 8:].columns[constant_filter.get_support()]]

train_features.iloc[:, 8:].drop(labels=constant_columns, axis=1, inplace=True)
test_features.iloc[:, 8:].drop(labels=constant_columns, axis=1, inplace=True)
```
